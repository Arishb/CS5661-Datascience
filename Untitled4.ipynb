```python
### Perform One hot encoding on people dataset
```

```python
###Fetures in people dataset on which One hot encoding can be performed
fc = ['char_1','char_2','char_3','char_4','char_5','char_6', 'char_7','char_8','char_9']
```

```python
### Logic for one hot encoding on people dataset
for column in fc :
    one_hot_enc = pd.get_dummies(df[column])
    # Drop column "activity_category" as it is now encoded
    df = df.drop(column, axis=1)
    # Join the encoded df
    df =df.join(one_hot_enc, lsuffix='_left', rsuffix='_right')
```

```python
#Create the feature matrix
X = df_new
```

```python
### Import library for PCA
from sklearn.decomposition import PCA
```

```python
### divide into testing and training set
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_new, test_size=0.25, random_state=2)
```

```python
### reduce dimensions. Use PCA
pca = PCA(n_components=10)
```

```python
pca.fit(X_train)
```

```python
X_transformed = pca.fit_transform(X_train)
X_transformed
```

```python
X_transformed = pd.DataFrame(X_transformed)
```

```python
X_transformed
```

```python
X_transformed_test = pca.fit_transform(X_test)
```

```python
X_transformed_test = pd.DataFrame(X_transformed_test)
```

```python
X_transformed_test
```

```python
##### Step13 : Import RandomForest classifier
```

```python
# The following line will import RandomForest Classes
# RandomForest Classifier" is the name of a "sklearn class" to perform Decision Tree based Classification

from sklearn.ensemble import RandomForestClassifier
```

```python
#### Step14 : Instantiate an "object" of RandomForestClassifier "class"
```

```python
my_RandomForest = RandomForestClassifier(n_estimators=19,bootstrap=True,random_state=2 )
```

```python
my_RandomForest.fit(X_transformed,y_train)
```

```python
y_predict_RF = my_RandomForest.predict(X_transformed_test)
```

```python
print(y_predict_RF)
```

```python
# Function "accuracy_score" from "sklearn.metrics" will perform the element-to-element comparision and returns the 
# portion of correct predictions:
from sklearn.metrics import accuracy_score
```

```python
score_RF = accuracy_score(y_test, y_predict_RF)
print(score_RF)
```
